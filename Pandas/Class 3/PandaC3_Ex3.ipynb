```python
import pandas as pd
from pathlib import Path
```

```python
#data1 =Path("C:/Users/sebat/ClassRepository/exercises_class/PANDAS/Class 3/Crowdfunding_data.csv")
data_path=Path("Resources/Crowdfunding_data.csv")
#data_path=Path("../Class 2/movie_score.csv")
```

```python
data_file=pd.read_csv(data_path)
data_file.head()
```

```python
data_file.columns
```

```python
#new_data=data_file[["name", "goal", "pledged", "outcome", "country", "staff_pick", "backers_count", "spotlight"]]
#Esto es para filtrar las columnas que yo necesito. (Es lo mismo que hace el codigo de arriba, solo que les gusta mas a los profe)
new_data=data_file.loc[:,["name", "goal", "pledged", "outcome", "country", "staff_pick", "backers_count", "spotlight"]]
new_data
```

```python
#Esto es para filtrar los que tienen 0 en presupuesto donado.
resume_newdata=new_data.loc[(new_data["pledged"]>0)]
resume_newdata
```

```python
columns=["name", "goal", "pledged", "outcome", "country", "staff_pick", "backers_count", "spotlight"]
#aca filtro todos las filas que tienen como pais US, y le pido las columnas.
resume_newdata2=resume_newdata.loc[(resume_newdata["country"]=="US",columns)]
resume_newdata2
```

```python
#Calculo el promedio de donación y ademas creo una nueva columna con este dato
resume_newdata2["Average Donation"]=resume_newdata2["pledged"]/resume_newdata2["backers_count"]
resume_newdata2
```

```python
resume_newdata2["Average Donation"]=resume_newdata2["Average Donation"].astype(float).map("${:,.2f}".format)
#resume_newdata2["Average Amount Pledged"]=resume_newdata2["Average Amount Pledged"].astype(float).map("${:,.2f}".format)
resume_newdata2["goal"]=resume_newdata2["goal"].astype(float).map("${:,.2f}".format)
resume_newdata2["pledged"]=resume_newdata2["pledged"].astype(float).map("${:,.2f}".format)
resume_newdata2
```
