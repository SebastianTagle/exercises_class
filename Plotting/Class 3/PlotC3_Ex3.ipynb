```python
 # Dependencies
import pandas as pd
import sklearn.datasets as dta
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
 # Read in the wine recognition data set from sklearn and load into Pandas
data = dta.load_wine()
wine_data = pd.DataFrame(data.data,columns=data.feature_names)
print(data.DESCR)
```

```python
wine_data.columns
```

```python
 # Plot flavanoids versus malic_acid on a scatterplot
plt.scatter(wine_data["flavanoids"],wine_data["malic_acid"])
plt.xlabel("flavanoids")
plt.ylabel("Malic acid")
plt.show()
```

```python
correlation=st.pearsonr(wine_data["flavanoids"],wine_data["malic_acid"])
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
plt.scatter(wine_data["alcohol"],wine_data["color_intensity"])
plt.ylabel("Colour Intensity")
plt.xlabel("Alcohol")
plt.show()
```

```python
correlation2=st.pearsonr(wine_data["color_intensity"],wine_data["alcohol"])
#hay que poner [0] para que devuelva el primera valor que calcula (calcula la correlación y el pvalue.)
print(f"The correlation between both factors is {round(correlation2[0],2)}")
```

```python
 # BONUS: Generate the correlation matrix and find the strongest positive and negative correlations
#calculate correlation the everthing vs everything
wine_corr = wine_data.corr()

wine_corr.unstack().sort_values()
```
