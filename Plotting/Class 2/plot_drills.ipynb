```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(0, len(gyms))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)

plt.title("NYC Gym Popularity")
plt.xlabel("Gym Name")
plt.ylabel("Number of Members")

# Finish the plot
plt.xlim(-.75,len(gyms)-.25)
plt.ylim(0,max(members)+5)

plt.bar(x_axis,members,facecolor="red",alpha=0.2,align="center")
plt.xticks(tick_locations,gyms)
plt.show()
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python

plt.title("Sin from 0 to 2$\pi$")
plt.xlabel("Real Numbers from 0 to 2$\pi$")
plt.ylabel("sin(x)")

# Finish the plot


plt.hlines(0,0,x_lim,alpha=0.2)
plt.xlim(0,x_lim)
plt.ylim(-1.25,1.25)

plt.plot(x_axis,sin,marker="^",color="red",linewidth=1)
plt.show()
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.title("NYC Gym Popularity")
plt.pie(members,explode=explode,labels=gyms, colors=colors, autopct="%1.1f%%",shadow=True,startangle=90)
plt.axis("equal")
plt.show()
# Finish the plot
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.title("Running Time of FakeSort for Sample Input Sizes")
plt.xlabel("Length of Input Array")
plt.ylabel("Time to Sort (s)")

# Finish the plot
plt.scatter(x_axis,times,marker="x",color="red")
plt.show()
```
