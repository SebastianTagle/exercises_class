```python
 # Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("union_settlements_1995.csv")
data.head()
```

```python
data.columns
```

```python
union_data=data["UNION"].value_counts()
union_data
```

```python
#detalle de parametros
#(Tipo de grafico, color del grafico, tamaÃ±o de grafico ancho y alto, titulo, nombre ejex y nombre eje y)
fig_1=union_data.plot(kind="bar",facecolor="blue",figsize=(8,6),
                     title="Major Collective Bargaing Settlement (1995)",
                     xlabel="Union",
                     ylabel="Settlement")

#agregar detalles al eje x
xticklabels=union_data.index
#nombre de cada columna, wrap=contrae los nombres de las columnas
fig_1.set_xticklabels(xticklabels, rotation=45, rotation_mode="anchor",ha="right",wrap=True)

plt.show()
```
